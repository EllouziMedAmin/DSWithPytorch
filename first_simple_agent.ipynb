```python
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

# Load Mistral-7B-Instruct
model_name = "mistralai/Mistral-7B-Instruct"
```

```python
from huggingface_hub import login
login("token")
```

```python
# Load the tokenizer and model
print("Loading model... (This may take a while)")
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, torch_dtype="auto", device_map="auto")
```

```python
# Create a text-generation pipeline
llm = pipeline("text-generation", model=model, tokenizer=tokenizer, max_new_tokens=150)
```

```python
def chat_with_agent():
    print("AI Agent: Hello! Ask me anything (type 'exit' to quit).")
    chat_history = []

    while True:
        user_input = input("You: ")
        if user_input.lower() == "exit":
            print("AI Agent: Goodbye!")
            break

        # Add user input to chat history
        chat_history.append(f"User: {user_input}")

        # Construct context properly
        context = "\n".join(chat_history) + "\nAI:"

        # Generate response (FIXED SETTINGS)
        response = llm(
            context,
            max_new_tokens=100,
            do_sample=True,  # Ensures `temperature` and `top_p` are used
            temperature=0.7,
            top_p=0.9
        )[0]['generated_text'].split("AI:")[-1].strip()

        # Print response
        print("AI Agent:", response)

        # Append AI response to history
        chat_history.append(f"AI: {response}")
```

```python
chat_with_agent()
```
